```python
#不要 refresh webdriver
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import random, time
from getcomment import getComment

login_url = 'https://www.jkforum.net/member.php?mod=logging&action=login&referer=https%3A%2F%2Fwww.jkforum.net%2Fhome%2F'
driver = webdriver.Chrome('chromedriver.exe')
driver.maximize_window()
driver.get(login_url)
input('登入完成請按 Enter')
root_xpath = '//*[@id="__layout"]/div/section/div[2]/section/div/div[2]/div[2]/'
author_xpath = '/div/div/div[1]/div[1]/div[1]/div[1]/div[2]/div'
content_xpath = '/div/div/div[1]/div[1]/div[2]/span'
title_xpath = '/div/div/div[1]/div[1]/div[3]/div/div/div/div/div/div/div[3]'
like_xpath = '/div/div/div[1]/div[2]/button[1]'
comment_xpath = '/div/div/div[1]/div[2]/button[2]'
text_xpath = '/div/div/div[2]/div/div/div/div/div[2]/textarea'
submit_xpath = '/div/div/div[2]/div[2]/div/div/div/div[2]/i'

amount = int(input('要留幾篇'))
print('倒數三秒後開始')
time.sleep(3)
for i in range(amount):
    #generate xpaths
    post_xpath = root_xpath + ('div[{}]').format(i+1)
    post_author_xpath = post_xpath + author_xpath
    post_content_xpath = post_xpath + content_xpath
    post_title_xpath = post_xpath + title_xpath
    post_like_xpath = post_xpath + like_xpath
    post_comment_xpath = post_xpath + comment_xpath
    post_text_xpath = post_xpath + text_xpath
    post_submit_xpath = post_xpath + submit_xpath
    
    #print info & generate comments
    post_index = '【第{}篇】'.format(i+1)
    post_author = driver.find_element_by_xpath(post_author_xpath).text
    post_content = driver.find_element_by_xpath(post_content_xpath).text
    try:
        post_title = driver.find_element_by_xpath(post_title_xpath).text
    except:
        post_content = post_content
    else:
        post_content = post_content + '\n文章標題:\n' + post_title
    
    post_comment = getComment(post_author, post_content)
    
    print(post_index, post_author, post_content, sep='\n', end='\n\n')
    while True:
        print('【留言】', post_comment)
        confirm = input('【OK>Enter】  【手動更改>輸入1】  【重新隨機>輸入任何字元】')
        if confirm == '':
            break
        elif confirm == '1':
            post_comment = input('請輸入留言： ')
        else:
            post_comment = getComment(post_author, post_content)
        
    #commenting
    like_button = driver.find_element_by_xpath(post_like_xpath)
    comment_button = driver.find_element_by_xpath(post_comment_xpath)
    if post_comment != '':
        like_button.click()
        comment_button.click()
        text_area = driver.find_element_by_xpath(post_text_xpath)
        text_area.send_keys(post_comment)
        text_area.send_keys(Keys.RETURN)
        print('【已留言】\n')
    else:
        like_button.click()
```
